##Transformación a DataFrame


```python
import pandas as pd

# leer lo que contiene ventas, marketing,cliente

df_ventas = pd.read_csv('ventas.csv')
df_clientes = pd.read_csv('clientes.csv')
df_marketing = pd.read_csv('marketing.csv')

print("¡Éxito! Los 3 sets de datos ya están en la memoria de Python.")
```

```python
# 1. Cargamos el archivo ventas (le indicamos que la columna 'fecha_venta' es una fecha)
df_ventas = pd.read_csv('ventas.csv', parse_dates=['fecha_venta'], dayfirst=True)

# 2.Limpieza del precio: quitamos el '$' y lo convertimos a número decimal (float)

df_ventas['precio'] = df_ventas['precio'].replace(r'[\$,]', '', regex=True).astype(float)
print("--- Tipos de Datos en Ventas ---")
print(df_ventas.dtypes)
```

```python
df_ventas.head()
```

```python
# Corregir los encabezados

df_ventas.columns = ['ID_Venta', 'Producto', 'Precio_Unitario', 'Cantidad', 'Fecha_venta', 'Categoria']

print("Encabezados aplicados correctamente:")
df_ventas.head()
```

```python
import pandas as pd


# 1. Cargamos el archivo indicando las dos columnas de fecha # Usamos dayfirst=True porque el formato es DD/MM/YYYY

df_marketing = pd.read_csv('marketing.csv', parse_dates=['fecha_inicio', 'fecha_fin'], dayfirst=True)

 # 2. Aplicamos para dar formato a encabezados

df_marketing.columns = ['ID_Campania', 'Producto', 'Canal', 'Costo', 'Fecha_Inicio', 'Fecha_Fin']

 # 3. Verificando

print("Estructura de Marketing:")
df_marketing.info()
df_marketing.head()
print("Estructura de Marketing finalizada:")







```

```python
import pandas as pd
df_clientes = pd.read_csv('clientes.csv')

#  Aplicar encabezados
df_clientes.columns = ['ID_Cliente', 'Nombre', 'Edad', 'Ciudad', 'Ingresos']

#  Verificando
print("Base de Clientes lista:")
df_clientes.head()

# formato a la columna Ingresos
df_clientes['Ingresos'] = df_clientes['Ingresos'].astype(float)

# Verificamos el tipo de dato
print(df_clientes['Ingresos'].dtype)

df_clientes.head()


```

##Lista de Diccionarios

```python
# 1. Convertimos los datos del DataFrame real a una Lista de Diccionarios
# Esto cumple con la consigna de usar estructuras de datos de Python puro
ventas_lista = df_ventas.to_dict('records')

# 2. Script para calcular ventas mensuales
# Usamos un diccionario para acumular los totales por mes
ventas_por_mes = {}

for venta in ventas_lista:
    # Extraemos el mes de la fecha (que ya habiamos limpiado antes)
    # .month_name() nos da el nombre del mes
    mes = venta['Fecha_venta'].month_name()

# Variables y Operadores

    cant = venta['Cantidad']
    prec = venta['Precio_Unitario']
    subtotal = cant * prec

    # Acumulación mensual
    if mes in ventas_por_mes:
        ventas_por_mes[mes] += subtotal
    else:
        ventas_por_mes[mes] = subtotal

# 4. Resultados finales
print("--- RESULTADOS MENSUALES ---")
for mes, total in ventas_por_mes.items():
    print(f"Mes: {mes:<12} | Total: ${total:,.2f}")
```

```python
# Reposte de nulos y duplicados
import pandas as pd

# Creamos una lista con nuestros DataFrames y sus nombres
datasets = {
    "Ventas": df_ventas,
    "Marketing": df_marketing,
    "Clientes": df_clientes
}

print("--- REPORTE DE CALIDAD DE DATOS (ESTADO INICIAL) ---\n")

for nombre, df in datasets.items():
    print(f"FICHA TÉCNICA: {nombre}")
    print(f"Total de registros: {len(df)}")

    # Identificar Nulos
    nulos = df.isnull().sum().sum()
    print(f"Valores nulos encontrados: {nulos}")
    if nulos > 0:
        print("Detalle de nulos por columna:")
        print(df.isnull().sum()[df.isnull().sum() > 0])

    # Identificar Duplicados
    duplicados = df.duplicated().sum()
    print(f"Registros duplicados: {duplicados}")

    print("-" * 30)
```

```python
import pandas as pd

# 1. Aseguramos que los Ingresos de clientes sean float
df_clientes['Ingresos'] = df_clientes['Ingresos'].astype(float)

# 2. Pasamos Ventas a Lista de Diccionarios (cumpliendo la consigna)
ventas_lista = df_ventas.to_dict('records')

# 3. Script de Cálculo Mensual con Operadores y Variables
ventas_por_mes = {}

for venta in ventas_lista:
    # Extraemos el mes
    mes = venta['Fecha_venta'].month_name()

    # Variables y Operadores (Alineación corregida)
    # Importante: Usamos 'Cantidad' y 'Precio_Unitario' con mayúsculas como tus encabezados
    cant = venta['Cantidad']
    prec = venta['Precio_Unitario']
    subtotal = cant * prec

    # Acumulación mensual
    if mes in ventas_por_mes:
        ventas_por_mes[mes] += subtotal
    else:
        ventas_por_mes[mes] = subtotal

# 4. Resultados finales
print("--- RESULTADOS MENSUALES ---")
for mes, total in ventas_por_mes.items():
    print(f"Mes: {mes:<12} | Total: ${total:,.2f}")
```

```python
# 1. Primero, calculamos el Total de cada fila
df_ventas['Total_Venta'] = df_ventas['Precio_Unitario'] * df_ventas['Cantidad']

# 2. Definimos el umbral de "Alto Rendimiento"
# En este caso, buscaremos ventas individuales mayores a $500 (ajusta el número según tus datos)
umbral = 500

# 3. Aplicamos la Transformación (Filtro)
df_alto_rendimiento = df_ventas[df_ventas['Total_Venta'] > umbral].copy()

# 4. Ordenamos de mayor a menor para que sea más visual
df_alto_rendimiento = df_alto_rendimiento.sort_values(by='Total_Venta', ascending=False)

print(f"Se identificaron {len(df_alto_rendimiento)} registros de alto rendimiento.")
df_alto_rendimiento.head()






```

##Agrupación por Producto

```python
# Agrupamos por producto y sumamos sus totales
ranking_productos = df_ventas.groupby('Producto')['Total_Venta'].sum().reset_index()

# Filtramos los productos que en total vendieron más de, por ejemplo, $5000
productos_top = ranking_productos[ranking_productos['Total_Venta'] > 5000]

print("Productos con alto rendimiento acumulado:")
print(productos_top)
```

```python
# 1. Aseguramos que exista la columna de Total por cada fila
df_ventas['Total_Venta'] = df_ventas['Precio_Unitario'] * df_ventas['Cantidad']

# 2. Definimos un umbral: Ventas que superen el promedio
promedio_ventas = df_ventas['Total_Venta'].mean()

# 3. Creamos el nuevo DataFrame con los productos de alto rendimiento
df_alto_rendimiento = df_ventas[df_ventas['Total_Venta'] > promedio_ventas].copy()

# 4. Ordenamos para ver los más importantes arriba
df_alto_rendimiento = df_alto_rendimiento.sort_values(by='Total_Venta', ascending=False)

print(f"Umbral de rendimiento aplicado: ${promedio_ventas:.2f}")
print(f"Registros que superan el promedio: {len(df_alto_rendimiento)}")
df_alto_rendimiento.head()
```
